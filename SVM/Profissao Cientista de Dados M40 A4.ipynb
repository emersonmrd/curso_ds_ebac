```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
base = pd.read_csv('fetal_health.csv')
```

```python
X = base.drop(columns=['fetal_health'])
Y = base['fetal_health']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

image.png

Imagine que voc tem dois grupos de pontos em um grfico 2D: tringulos (A) e crculos (B). Eles esto bem separados, e voc pode traar uma linha reta entre eles para dividir as classes. O SVM com kernel linear encontra essa linha que melhor separa os dois grupos, garantindo a maior distncia possvel entre a linha e os pontos mais prximos de cada grupo. Assim, ele pode classificar novos pontos dependendo de qual lado da linha eles caem.

```python
svm_model = SVC(kernel =  'linear', random_state = 1, C = 1.0)
svm_model.fit(X_train,Y_train)
```

```python
Y_pred = svm_model.predict(X_train)
Y_pred
```

```python
report = classification_report(Y_train, Y_pred)
print("Relatrio de Classificao:\n", report)
```

```python
accuracy = accuracy_score(Y_train, Y_pred)
print("Acurcia:", accuracy)
```

# Melhorando os parmetros:

image.png

O kernel polinomial permite ao SVM operar como se tivesse criado novas variveis (ou caractersticas) no espao de maior dimenso, possibilitando a separao de dados que no podem ser separados linearmente no espao original.

```python
svm_model_poly = SVC(kernel =  'poly', random_state = 1, C = 1.0)
svm_model_poly.fit(X_train,Y_train)
```

```python
Y_pred_poly = svm_model_poly.predict(X_train)
Y_pred_poly
```

```python
report = classification_report(Y_train, Y_pred_poly)
print("Relatrio de Classificao:\n", report)
```

```python
accuracy = accuracy_score(Y_train, Y_pred_poly)
print("Acurcia:", accuracy)
```

image.png

```python
svm_model_rbf = SVC(kernel =  'rbf', random_state = 1, C = 1.0)
svm_model_rbf.fit(X_train,Y_train)
```

```python
Y_pred_rbf = svm_model_rbf.predict(X_train)
Y_pred_rbf
```

```python
report = classification_report(Y_train, Y_pred_rbf)
print("Relatrio de Classificao:\n", report)
```

```python
accuracy = accuracy_score(Y_train, Y_pred_rbf)
print("Acurcia:", accuracy)
```

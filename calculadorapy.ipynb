```python
def principal():
    # Escolha do primeiro número
    num1 = escolha_primeiro()

    # Escolha do segundo número
    num2 = escolha_segundo()

    # Informa os valores digitados pelo usuário
    print(f"Você digitou {num1} e {num2}")

    # Pergunta qual o operador o usuário deseja utilizar
    print("Qual operação você deseja fazer? ")

    # Escolha do operador
    operador = escolha_operadores()

    if operador == 1:
        sinal = "+"
    elif operador == 2:
        sinal = "-"
    elif operador == 3:
        sinal = "x"
    elif operador == 4:
        sinal = "/"

    # Função para calcular
    resultado = calcula(num1, num2, operador)

    # Mostra o resultado da operação para o usuário
    print()
    print("Expressão: ")
    print(f"{num1} {sinal} {num2} = {resultado}")

def cabecalho():
    # CABEÇALHO DA APLICAÇÃO
    print("-" * 30)
    print("CALCULADORA PY".center(30, " "))
    print("-" * 30)
    print()

# Função para escolher o primeiro número
def escolha_primeiro():
    while True:
        try:
            num1 = float(input("Digite um número: "))
            return num1
        except ValueError:
            print("Você não digitou um número, tente novamente!")

# Função para escolher o segundo número
def escolha_segundo():
    while True:
        try:
            num2 = float(input("Digite outro número: "))
            return num2
        except ValueError:
            print("Você não digitou um número, tente novamente!")

# Função para escolher o operador
def escolha_operadores():
    while True:
        operador = input("""
        Escolha uma operação:
        1: +
        2: -
        3: x
        4: /

        Opção: """)
        if operador.isdigit() and operador in ['1', '2', '3', '4']:
            return int(operador)
        else:
            print("Essa opção não é válida, tente novamente!")

# Função para calcular a operação
def calcula(valor1, valor2, operacao):
    if operacao == 1:
        return valor1 + valor2
    elif operacao == 2:
        return valor1 - valor2
    elif operacao == 3:
        return valor1 * valor2
    elif operacao == 4:
        return valor1 / valor2

# Executa o cabeçalho
cabecalho()

resposta = "S"
while resposta != "N":
    principal()

    # Pergunta se o usuário deseja continuar
    print()
    resposta = input("Você deseja continuar? [S/N] ").upper().strip()
    print()
    while resposta not in "SN":
        print("Por favor responda S ou N")
        resposta = input("Você deseja continuar? [S/N] ").upper().strip()
        print()

print("SAINDO... ATÉ LOGO!")
```
